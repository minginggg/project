* 음성  &nbsp;- &nbsp;  네이버 클로바 케어콜 NEST

여기서 NEST는 제한된 데이터 학습만으로도, 복잡하고 다양한 장문의 의미,음성 표현을 정확하게 인식하고, 텍스트로 변환할 수 있는 기술입니다.대량의 정제된 데이터를 사전에 학습하지 않고도, 예상치 못한 표현에 대해 정확한 음성인식이 가능한 것이 장점입니다. 음향 정보와 언어 정보를 별도로 학습하는 기존의 모델링 방식을 통합 모델링 방식(end-to-end)으로 개선해, 학습에 필요한 데이터의 양과 시간은 기존의 1/10 수준으로 단축시키면서도, 인식의 정확도는 오히려 높였습니다.

* 자율주행  &nbsp;- &nbsp;  기아 자동차 scc-ml

  &nbsp; &nbsp; SCC-ML의 원리는 먼저 전방카메라, 레이더 등의 센서가 다양한 운전상황에서 발생되는 정보를 지속적으로 수집해 ADAS(첨단 운전자 지원 시스템)의 두뇌격인 제어컴퓨터로 보냅니다. 제어컴퓨터는 입력된 정보로부터 운전자의 주행습관을 추출해 종합적인 주행성향을 파악하고 이때 인공지능 기술 중 하나인 머신러닝 알고리즘이 적용됩니다.
  스스로 학습한 인공지능은 SCC 작동 때 운전자와 거의 흡사한 패턴으로 자율주행을 지원해줍니다.주행성향에 대한 정보는 센서를 통해 계속 업데이트되기 때문에 운전자의 최근 성향을 반영할 수 있고 안전운전을 크게 벗어난 주행성향은 따르지 않도록 설정돼 있어서 신뢰성을 높였습니다.

  조사를 하면서 주변에 인공지능이 적용된 제품이나 서비스가 꽤 많다는 걸 알게 됐습니다. 특히 이번 코로나 사태에서 인공지능이 적용된 사례가 있다는 건 더 놀라웠고 조사한 4가지 중 음성 케어콜은 정말 관심이 많이 갔습니다. 배우고 싶어했던 헬스케어 분야에도 케어콜과 같은 서비스가 필요할 것이라고 생각합니다.
그래서 케어콜처럼 고객이 질문을 하면 답을 내릴 수 있는 인공지능을 처음엔 공부해 나가고 싶습니다. 그 이후 헬스케어 분야와 조합할 수 있도록 개발하는 것이 목표입니다.
